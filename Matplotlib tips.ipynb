```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 50, 40)
ax.plot(x, np.sin(x), '-oc', label='sin(x)')
ax.plot(x, np.cos(x), color='red', linestyle='--', label='cos(x)')
ax.plot(x, np.cos(x + 2), color='orange', linestyle='-.', label='cos(x + 2)')
ax.plot(x, np.cos(x - 1), color='g', linestyle=':', label='cos(x - 1)')
ax.plot(x, np.arctan(x), color='green' , label='arctan(x)')
plt.legend()

plt.axis([-1, 25, -1.5, 2])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')

x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.gaussian_process import GaussianProcess

x = np.linspace(0, 10, 50)

model = lambda x: x * np.sin(x)

xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1e-1, random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE)

plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color='gray', alpha=0.2)
plt.xlim(0, 10)


```
